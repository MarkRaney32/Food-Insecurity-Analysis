# Plotting Census Tracts

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.figure_factory as ff
import geopandas as gpd
import plotly.express as px
import math
from lxml import etree
from matplotlib.pyplot import figure
from matplotlib.colors import ListedColormap
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# Reading the dataset containing polygons of census tracts.
# Modify the read_file parameter to fit the path of
# where the file is located.

geoData = gpd.read_file("tl_2019_39_tract/tl_2019_39_tract.shp")
geoData['GEOID'] = geoData['GEOID'].astype('int64')
geoData = geoData.rename(columns = {'GEOID':'CensusTract'})
```

```python
# Fetching the final metrics that we generated
ruralFinal = pd.read_csv('rural_our_FINAL_FI.csv')
urbanFinal = pd.read_csv('urban_our_FINAL_FI.csv')
```

```python
# Merging polygons and final metrics along tract id
ruralGeo = ruralFinal.merge(geoData[['CensusTract', 'geometry']], on='CensusTract')
urbanGeo = urbanFinal.merge(geoData[['CensusTract', 'geometry']], on='CensusTract')
ruralGeo = gpd.GeoDataFrame(ruralGeo)
urbanGeo = gpd.GeoDataFrame(urbanGeo)
```

```python
# Plotting Rural Tracts - Coloring by final food insecurity score
fig, ax = plt.subplots(figsize=(13, 13))
ruralGeo.plot(ax=ax, column = 'final_FI_Score', cmap = plt.get_cmap('Reds'), legend = True)

# Plotting cities as scatter points just for reference
cities = [[-82.9988, -81.6944, -84.5120, -83.5379, -84.1916],
          [39.9612,   41.4993,  39.1031,  41.6528,  39.7589]]
cityNames = ['Columbus', 'Cleveland', 'Cincinnati', 'Toledo', 'Dayton']
ax.scatter(cities[0], cities[1], color = 'black')
for i in range(len(cityNames)):
    ax.text(cities[0][i] + 0.05, cities[1][i] + 0.05, cityNames[i])

#Labeling graph
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_title("Food insecurity rating by tract - Rural")

# Uncomment below if you want to same image to working directory
#plt.savefig('rural.png')
```
